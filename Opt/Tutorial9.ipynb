# Question 1

For each of these functions, write down each of their partial derivitives

a) $$ f(x,y) = x^2+3xy−y^2 $$

$$\frac{\delta f(x, y)}{\delta x} = 2x + 3y$$
$$\frac{\delta f(x, y)}{\delta y} = 3x - 2y$$

b) $$ f(x,y) = cos(xy) $$

$$\frac{\delta f(x, y)}{\delta x} = \frac{\delta cos(xy)}{\delta xy} \cdot \frac{\delta xy}{\delta x} = -sin(x, y) \cdot y$$

$$\frac{\delta f(x, y)}{\delta y} = \frac{\delta cos(xy)}{\delta xy} \cdot \frac{\delta xy}{\delta y} = -sin(x, y) \cdot x$$

c) $$ f(x,y) = (1-x)^2 + 5*(y - x^2)^2 $$

$$ \frac{\delta f(x, y)}{\delta x} 
= \frac{\delta ( (1 - x)^2 + 5 * (y - x^2)^2 )}{\delta x} 
= \frac{\delta (1 - x)^2}{\delta x} + \frac{\delta 5 * (y - x^2)^2}{\delta x} $$
$$ = \frac{\delta (1 - x)^2}{\delta (1 - x)} \cdot \frac{\delta (1 - x)}{\delta x} + \frac{\delta 5 * (y - x^2)^2}{\delta (y - x^2)^2} \cdot \frac{\delta (y - x^2)^2}{\delta (y - x^2)} \cdot \frac{\delta (y - x^2)}{\delta x} $$
$$ = -2(1 - x) + 5 \cdot 2 \cdot (y - x^2) \cdot (-2x) = -2 + 2x -20xy + 20x^3 $$

$$ \frac{\delta f(x, y)}{\delta y} = \frac{\delta ( (1 - x)^2 + 5 * (y - x^2)^2 )}{\delta y} = \frac{\delta (1 - x)^2}{\delta y} + \frac{\delta 5 \cdot (y - x^2)^2}{\delta y} $$
$$ = 0 + 5 \cdot \frac{\delta (y - x^2)^2}{\delta (y - x^2)} \cdot \frac{\delta (y - x^2)}{\delta y} = 10 \cdot (y - x^2) $$

d) $$ f(x,y) = x^2 + y^2 + x*sin(y) $$

$$ \frac{\delta f(x, y)}{\delta x} = \frac{\delta(x^2 + y^2 + x \cdot sin(y))}{\delta x} = \frac{\delta x^2}{\delta x} + \frac{\delta y^2}{\delta x} + \frac{\delta x \cdot sin(y)}{\delta x} = 2x + sin(y) $$

$$ \frac{\delta f(x, y)}{\delta y} = \frac{\delta(x^2 + y^2 + x \cdot sin(y))}{\delta y} = \frac{\delta x^2}{\delta y} + \frac{\delta y^2}{\delta y} + \frac{\delta x \cdot sin(y)}{\delta y} = 2y + x cos(y) $$
